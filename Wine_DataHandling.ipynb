```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('df_wine_all.csv')
```

```python
train = df = data[data['alcohol'].notna()]
```

```python
# price
pc = []
price = train['price']
for p in price:
    if p != None:
        p = p.replace('£', '')
        pc.append(float(p))
    else:
        pc.append(None)
train['price'] = pc
```

```python
# vintage
vn = []
vintage = train['vintage']
for v in vintage:
    if v != None and v != np.nan:
        vn.append(float(v))
    else:
        vn.append(None)
train['vintage'] = vn
```

```python
# country
train['country'].value_counts()
country = train['country']
cc = []
for c in country:
    if c == 'France':
        cc.append(1)
    elif c =='Italy':
        cc.append(2)
    elif c == 'Spain':
        cc.append(3)
    elif c == 'South Africa':
        cc.append(4)
    elif c == 'Australia':
        cc.append(5)
    elif c == 'Chile':
        cc.append(6)
    elif c == 'USA':
        cc.append(7)
    else:
        cc.append(8)
train['country'] = cc

```

```python
# types
train['type'].value_counts()
types = train['type']
tt = []
for c in types:
    if c == 'Still Red Wine':
        tt.append(1)
    elif c =='Still White Wine':
        tt.append(2)
    elif c == 'Sparkling Wine':
        tt.append(3)
    elif c == 'Rose':
        tt.append(4)
    elif c == 'Sweet/Dessert':
        tt.append(5)
    else:
        tt.append(6)
train['type'] = tt
```

```python
# grapespecies
train['grapespecies'].value_counts()
grapespeciess = train['grapespecies']
gr = []
for c in grapespeciess:
    if c == '100% Pinot Noir ':
        gr.append(1)
    elif c =='100% Chardonnay':
        gr.append(2)
    elif c == '100% Sauvignon Blanc':
        gr.append(3)
    elif c == 'Chardonnay':
        gr.append(4)
    elif c == '100% Malbec':
        gr.append(5)
    elif c == '100% Merlot':
        gr.append(6)
    elif c == '100% Riesling':
        gr.append(7)
    elif c == '100% Shiraz':
        gr.append(8)
    else:
        gr.append(9)
train['grapespecies'] = gr
```

```python
# region        
train['region'].value_counts()
regions = train['region']
re = []
for c in regions:
    if c == 'Burgundy':
        re.append(1)
    elif c =='Languedoc':
        re.append(2)
    elif c == 'Veneto':
        re.append(3)
    elif c == 'Piemonte':
        re.append(4)
    elif c == 'Rhone':
        re.append(5)
    elif c == 'Mendoza':
        re.append(6)
    elif c == 'Rioja':
        re.append(7)
    elif c == 'Rioja':
        re.append(8)
    elif c == 'Western Cape':
        re.append(9)
    else:
        re.append(10)
train['region'] = re
```

```python
# bottle size
bs = []
bottle = train['bottle size']
for b in bottle:
    if b != None:
        b = b.replace('cl', '')
        bs.append(float(b))
    else:
        bs.append(None)
train['bottle size'] = bs
```

```python
#alcohol
al = []
alcohol = train['alcohol']
train['alcohol']
for a in alcohol:
    if a != None:
        a = a.replace('%','')
        al.append(float(a))
    else:
        al.append(None)
train['alcohol'] = al
```

```python
train.duplicated().sum()
train[train.duplicated()]
train.drop_duplicates(inplace=True)
```

```python
train = train.dropna()
train
```

```python
#outliers

from pandas.api.types import is_numeric_dtype
import numpy as np

cols=train[train.columns]##outliers
for i in cols:
    if is_numeric_dtype(train[i]):
        z_score=(train[i]-train[i].mean())/train[i].std()
        outliers=(abs(z_score)>3)
        train[i][outliers]=np.nan 
        
train = train.dropna()
train
```

```python
#Remove punctuation 
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string
    translator = str.maketrans('', '', string.punctuation) 
    return text.translate(translator)
```

```python
train['Text wine'] = train['Text wine'].apply(remove_punctuation)
```

```python
#import nltk      
#nltk.download() #run only in first time 
from nltk.corpus import stopwords
import numpy as np
# extracting the stopwords from nltk library
sw = stopwords.words('english')
# displaying the stopwords
#np.array(sw)
```

```python
def stopwords(text):
    '''a function for removing the stopword'''
    # removing the stop words and lowercasing the selected words
    text = [word.lower() for word in text.split() if word.lower() not in sw]
    # joining the list of words with space separator
    return " ".join(text)
train['Text wine'] = train['Text wine'].apply(stopwords)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

count_vec = CountVectorizer(max_features=10)
count_vec.fit(train['Text wine'])
text_vectorized = pd.DataFrame(count_vec.transform(train['Text wine']).toarray(), columns=count_vec.get_feature_names())
text_vectorized = text_vectorized.drop(index = 3)
frames = [train, text_vectorized]
df_merge = pd.concat(frames , axis = "columns")
text_vectorized
```

```python
df_merge = df_merge.dropna()
df_merge
```
